# Cleaning wBudget dataframe 

```python
import pandas as pd
import numpy as np 
import pickle
```

```python
fl_movies_wbudget = pd.read_pickle('fl_movies_wbudget.pickle')
```

```python
fl_movies_wbudget.drop(columns=['budget','currency','budget_new'], inplace=True)
```

```python
fl_movies_wbudget = fl_movies_wbudget[fl_movies_wbudget.budget_usd.notna()]
```

```python
fl_movies_wbudget.info()
```

```python
fl_movies_wbudget.max_theaters.fillna(fl_movies_wbudget.max_theaters.median(), inplace=True)
fl_movies_wbudget.domestic_opening.fillna(fl_movies_wbudget.domestic_opening.median(), inplace=True)
fl_movies_wbudget.num_opening_theaters.fillna(fl_movies_wbudget.num_opening_theaters.median(), inplace=True)
fl_movies_wbudget.runtime.fillna(fl_movies_wbudget.runtime.median(), inplace=True)
fl_movies_wbudget.earliest_release_location_opening_gross.fillna(fl_movies_wbudget.earliest_release_location_opening_gross.median(), inplace=True)
fl_movies_wbudget.earliest_release_location_original_gross.fillna(fl_movies_wbudget.earliest_release_location_original_gross.median(), inplace=True)
```

```python
fl_movies_wbudget.info()
```

```python
fl_movies_wbudget.describe()
```

```python
fl_movies_wbudget[fl_movies_wbudget['budget_usd'] =='Failed to convert.']
```

```python
fl_movies_wbudget = fl_movies_wbudget[fl_movies_wbudget.budget_usd != 'Failed to convert.']
```

```python
fl_movies_wbudget.info()
```

```python
fl_movies_wbudget['budget_usd'] = fl_movies_wbudget['budget_usd'].astype(float)
```

```python
fl_movies_wbudget['domestic_total_profit'] = fl_movies_wbudget.domestic_total_gross - (fl_movies_wbudget.budget_usd*(fl_movies_wbudget.domestic_total_gross/(fl_movies_wbudget.domestic_total_gross+fl_movies_wbudget.international_total_gross)))
```

```python
fl_movies_wbudget['international_total_profit'] = fl_movies_wbudget.international_total_gross - (fl_movies_wbudget.budget_usd*(fl_movies_wbudget.international_total_gross/(fl_movies_wbudget.domestic_total_gross+fl_movies_wbudget.international_total_gross)))
```

```python
fl_movies_wbudget.reset_index(inplace=True)
```

```python
fl_movies_wbudget.info()
```

```python
fl_movies_wbudget.describe()
```

```python
fl_movies_wbudget.to_pickle('fl_movies_wbudget2.pickle')
```

# Adjusting for inflation

```python
fl_movies_wbudget['year'] = pd.DatetimeIndex(fl_movies_wbudget['release_date']).year
```

```python
fl_movies_wbudget.info()
```

```python
fl_movies_wbudget.dropna(inplace=True)
```

```python
fl_movies_wbudget.info()
```

```python
import cpi 
cpi.update()
```

```python
def inflate_column(column):
    new_column = []
    for i in range(len(fl_movies_wbudget)):
        if fl_movies_wbudget['year'][i] > 2018:
            new_column.append(fl_movies_wbudget[column][i])
        else:
            new_column.append(cpi.inflate(fl_movies_wbudget[column][i],fl_movies_wbudget['year'][i]))
    return new_column
```

```python
fl_movies_wbudget['domestic_opening_real'] = inflate_column('domestic_opening')
```

```python
fl_movies_wbudget['domestic_total_gross_real'] = inflate_column('domestic_total_gross')
fl_movies_wbudget['international_total_gross_real'] = inflate_column('international_total_gross')
fl_movies_wbudget['earliest_release_location_opening_gross_real'] = inflate_column('earliest_release_location_opening_gross')
fl_movies_wbudget['earliest_release_location_original_gross_real'] = inflate_column('earliest_release_location_original_gross')
fl_movies_wbudget['budget_usd_real'] = inflate_column('budget_usd')
fl_movies_wbudget['domestic_total_profit_real'] = inflate_column('domestic_total_profit')
fl_movies_wbudget['international_total_profit_real'] = inflate_column('international_total_profit')
```

```python
fl_movies_wbudget.international_total_gross_real.describe()
```

```python
fl_movies_wbudget.international_total_gross.describe()
```

```python
fl_movies_wbudget.drop(columns=["domestic_total_gross", "international_total_gross", "earliest_release_location_opening_gross", "earliest_release_location_original_gross", "budget_usd", "domestic_total_profit", "international_total_profit"], inplace=True)
```

```python
fl_movies_wbudget.drop(columns=["domestic_opening","lifetime_gross"], inplace=True)
```
