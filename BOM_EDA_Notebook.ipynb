### Dropping duplicate columns

```python
import dateutil.parser

def money_to_int(moneystring):
    moneystring = moneystring.replace('$', '').replace(',', '')
    return int(moneystring)

def runtime_to_minutes(runtimestring):
    runtime = runtimestring.split()
    try:
        minutes = int(runtime[0])*60 + int(runtime[2])
        return minutes
    except:
        return None

def to_date(datestring):
    date = dateutil.parser.parse(datestring)
    return date
```

### Changing column datatypes

Changing release date to datetime object

```python
fl_movies['release_date_x']= pd.to_datetime(fl_movies['release_date_x'], format='%b %d, %Y')
```

Changing lifetime_gross and domestic_opening_x to money float objects

```python
fl_movies[['lifetime_gross']] = fl_movies[['lifetime_gross']].replace('[\$,]','',regex=True).astype(float)
```

```python
fl_movies.domestic_opening_x.replace({'-': None}, inplace=True)
```

```python
fl_movies[['domestic_opening_x']] = fl_movies[['domestic_opening_x']].replace('[\$,]','',regex=True).astype(float)
```

```python
fl_movies.info()
```

```python
fl_movies.describe()
```

Convert gross rank to float

```python
fl_movies['gross_rank'] = fl_movies['gross_rank'].str.replace(',', '').astype(float)
```

Convert max_theaters to float

```python
fl_movies.max_theaters.replace({'-': None}, inplace=True)
```

```python
fl_movies['max_theaters'] = fl_movies['max_theaters'].str.replace(',', '').astype(float)
```

Convert num_opening_theaters to float

```python
fl_movies.num_opening_theaters.replace({'-': None}, inplace=True)
```

```python
fl_movies['num_opening_theaters'] = fl_movies['num_opening_theaters'].str.replace(',', '').astype(float)
```

Replacing all "-" values with None 

```python
fl_movies.replace({'-': None}, inplace=True)
```

Removing "\n\n" from domestic distributor

```python
fl_movies.domestic_distributor_x.replace('\\n\\n', '', regex=True, inplace=True)
```

Changing a few column names

```python
fl_movies.rename(columns={"domestic_opening_x": "domestic_opening", "release_date_x": "release_date", "domestic_distributor_x": "domestic_distributor", "release_location":"earliest_release_location"}, inplace=True)
```
